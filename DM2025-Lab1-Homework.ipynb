```python
### Begin Assignment Here
# ===============================================================
# DM2025 - Lab1 - Homework (Phase 1, one-cell solution)
# Minimal comments are embedded below to explain WHAT/WHY briefly.
# ===============================================================

# ----------------------
# Imports
# ----------------------
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer

# ----------------------
# 1) Load the new dataset
#    - Goal: read `reddit-stock-sentiment.csv` from /newdataset (or fallbacks)
#    - Why: start Phase-1 using the *new* dataset (text + label).
# ----------------------
DATA_PATH = Path("newdataset") / "reddit-stock-sentiment.csv"   # ← adjust if your file lives elsewhere

# Try a few common fallback locations to reduce path issues in grading
if not DATA_PATH.exists():
    candidates = [
        Path.cwd() / "reddit-stock-sentiment.csv",
        Path.cwd().parent / "reddit-stock-sentiment.csv",
        Path("data") / "reddit-stock-sentiment.csv",
    ]
    for c in candidates:
        if c.exists():
            DATA_PATH = c
            break

# Fail fast with a clear message if the file still isn't found
if not DATA_PATH.exists():
    raise FileNotFoundError(
        f"Cannot find 'reddit-stock-sentiment.csv'. Please set DATA_PATH correctly. "
        f"Checked: {Path.cwd()}, {Path.cwd().parent}, ./data"
    )

# Be tolerant to weird rows; skip bad lines instead of crashing
df_raw = pd.read_csv(DATA_PATH, encoding="utf-8", engine="python", on_bad_lines="skip")

# ----------------------
# 2) Normalize columns to ['text', 'label']
#    - Goal: keep only the columns we need; coerce labels into {-1,0,1}
#    - Why: keeps the pipeline simple/consistent across datasets.
# ----------------------
text_candidates  = ["text", "body", "content", "message"]
label_candidates = ["label", "sentiment", "polarity", "score", "target"]

text_col  = next((c for c in text_candidates  if c in df_raw.columns), None)
label_col = next((c for c in label_candidates if c in df_raw.columns), None)

if text_col is None or label_col is None:
    raise ValueError(
        f"Could not find text/label columns. Found columns: {list(df_raw.columns)}. "
        f"Expected one of text={text_candidates} and label={label_candidates}."
    )

# Keep the two columns and rename uniformly
X = df_raw[[text_col, label_col]].rename(columns={text_col: "text", label_col: "label"}).copy()

# Minimal cleaning: remove empty text (prevents empty tokens later)
X["text"] = X["text"].astype(str).str.strip()
X = X[X["text"].str.len() > 0].reset_index(drop=True)

# If labels are strings (e.g., 'positive'), map to {-1,0,1}; otherwise coerce to numeric
if X["label"].dtype == "object":
    lbl_lower = X["label"].astype(str).str.lower().str.strip()
    map_guess = {
        "positive": 1, "pos": 1, "1": 1, "1.0": 1, "true": 1,
        "neutral": 0, "neu": 0, "0": 0, "0.0": 0,
        "negative": -1, "neg": -1, "-1": -1, "-1.0": -1, "false": -1,
    }
    X["label"] = lbl_lower.map(map_guess).fillna(np.nan)
X["label"] = pd.to_numeric(X["label"], errors="coerce")
X = X[X["label"].isin([-1, 0, 1])].reset_index(drop=True)   # keep valid labels only

# Quick sanity print (WHAT we loaded; WHY: verify schema/size)
print("=== Cleaned dataframe ===")
print(X.head(3))
print(f"Shape: {X.shape}")

# ----------------------
# 3) Inspect label distribution & show one sample per class
#    - WHAT: counts of -1/0/1; one short example each.
#    - WHY : understand imbalance early (affects modeling later).
# ----------------------
dist = X["label"].value_counts().sort_index()
print("\n=== Class distribution (label: count) ===")
for k, v in dist.items():
    print(f"{int(k)}: {v}")

print("\n=== One example per class ===")
for k in [-1, 0, 1]:
    if (X["label"] == k).any():
        ex = X[X["label"] == k].head(1)["text"].iloc[0]
        snippet = ex[:120].replace("\n", " ")  # precompute to avoid backslash in f-string
    else:
        snippet = "(none available)"
    print("Example for label={}: {} ...".format(k, snippet))

# ----------------------
# 4) Sampling + label distribution plots
#    - WHAT: sample subset for faster viz; compare distributions.
#    - WHY : check that sampling didn’t distort proportions.
# ----------------------
n_sample  = min(1000, len(X))                          # cap for readability
X_sample  = X.sample(n=n_sample, random_state=42).reset_index(drop=True)

def plot_label_distribution(df, title):
    counts = df["label"].value_counts().sort_index()
    plt.figure(figsize=(6, 3))
    plt.bar(counts.index.astype(str), counts.values)
    plt.title(title)
    plt.xlabel("label (-1/0/1)")
    plt.ylabel("count")
    ymax = counts.values.max()
    plt.ylim(0, ymax + max(10, int(0.05 * ymax)))
    plt.tight_layout()
    plt.show()

# Full dataset distribution
plot_label_distribution(X, "Category distribution (full data)")
# Sample distribution
plot_label_distribution(X_sample, "Category distribution (sample)")

# Side-by-side view to verify sample ≈ full
labels     = sorted(dist.index.tolist())
full_vals  = [dist.get(k, 0) for k in labels]
sample_vals= [int((X_sample["label"] == k).sum()) for k in labels]

x = np.arange(len(labels))
w = 0.35
plt.figure(figsize=(7, 3.5))
plt.bar(x - w/2, full_vals,   width=w, label="full")
plt.bar(x + w/2, sample_vals, width=w, label="sample")
plt.xticks(x, [str(int(k)) for k in labels])
plt.title("Category distribution (full vs sample)")
plt.xlabel("label")
plt.ylabel("count")
plt.legend()
plt.tight_layout()
plt.show()

# ----------------------
# 5) CountVectorizer (bag-of-words)
#    - WHAT: convert text → sparse doc-term matrix.
#    - WHY : inspect vocabulary, sparsity, and tokenization behavior.
# ----------------------
count_vect = CountVectorizer()                     # (could add stop_words='english' if desired)
X_counts  = count_vect.fit_transform(X["text"])    # sparse [n_docs, n_terms]

print("\n=== CountVectorizer ===")
print("Doc-term matrix shape:", X_counts.shape)    # large and sparse is expected

# Demo the default analyzer to illustrate tokenization effect
analyze      = count_vect.build_analyzer()
demo_tokens  = analyze("I am craving for a Hawaiian pizza right now!")
print("Analyzer demo tokens:", demo_tokens[:10], "...")

# Peek at a few vocabulary terms
feat_names = count_vect.get_feature_names_out()
print("First 5 terms in vocabulary:", feat_names[:5])

# Show non-zero terms for one doc (illustrates sparsity)
doc_idx  = min(4, X_counts.shape[0]-1)
nz_cols  = X_counts[doc_idx].nonzero()[1]
doc_terms= feat_names[nz_cols]
print(f"Non-zero terms in document index {doc_idx} (first 20):", list(doc_terms[:20]))

# ----------------------
# 6) Term frequency & visualizations
#    - WHAT: column sums = global term frequencies.
#    - WHY : show head vs long tail; log scale reveals mid/low-freq detail.
# ----------------------
term_freq = np.asarray(X_counts.sum(axis=0)).ravel()
order_desc = term_freq.argsort()[::-1]  # descending index (used by plots below)

def plot_top_k_terms(freq_vec, feature_names, k=200, title="Top-k term frequencies"):
    k     = min(k, len(feature_names))
    order = freq_vec.argsort()[::-1][:k]
    terms = feature_names[order]
    vals  = freq_vec[order]
    plt.figure(figsize=(14, 4))
    plt.bar(terms, vals)
    plt.title(f"{title} (k={k})")
    plt.ylabel("frequency")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

# Head (raw counts): very tall bars for a small set of terms
plot_top_k_terms(term_freq, feat_names, k=200, title="Top terms by raw frequency")

def plot_long_tail(freq_vec, feature_names, k=400, title="Long-tail frequency view"):
    order = freq_vec.argsort()[::-1]    # full sort; we’ll plot first k for readability
    terms = feature_names[order][:k]
    vals  = freq_vec[order][:k]
    plt.figure(figsize=(16, 4))
    plt.bar(terms, vals)
    plt.title(f"{title} (first {k})")
    plt.ylabel("frequency")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

# Long-tail (raw): shows steep head and very long flat tail
plot_long_tail(term_freq, feat_names, k=400, title="Long-tail frequency view (raw)")

# Log transform: compresses the head; mid/low-freq variation becomes visible
term_freq_log = np.log1p(term_freq)  # log(1+f) avoids log(0)
plot_top_k_terms(term_freq_log, feat_names, k=200, title="Top terms (log frequencies)")
plot_long_tail(term_freq_log, feat_names, k=400, title="Long-tail frequency view (log)")

print("\nDone. Phase-1 core steps completed successfully.")
```

```python
# =========================
# Phase 2
# Baseline: Count vs TF-IDF with MultinomialNB
# Booster : n-grams + vocabulary pruning + class weights + alpha search
# =========================
from pathlib import Path
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# ---------- 0) Load and normalize the dataset (build X=['text','label']) ----------
def _load_newdataset_as_X():
    DATA_PATH = Path("newdataset") / "reddit-stock-sentiment.csv"
    if not DATA_PATH.exists():
        # fallbacks to common locations
        for c in [Path.cwd() / "reddit-stock-sentiment.csv",
                  Path.cwd().parent / "reddit-stock-sentiment.csv",
                  Path("data") / "reddit-stock-sentiment.csv"]:
            if c.exists():
                DATA_PATH = c
                break
    if not DATA_PATH.exists():
        raise FileNotFoundError("Cannot find reddit-stock-sentiment.csv")

    df = pd.read_csv(DATA_PATH, encoding="utf-8", engine="python", on_bad_lines="skip")

    # normalize columns -> ['text','label']
    text_cands  = ["text", "body", "content", "message"]
    label_cands = ["label", "sentiment", "polarity", "score", "target"]
    tcol = next((c for c in text_cands  if c in df.columns), None)
    lcol = next((c for c in label_cands if c in df.columns), None)
    if tcol is None or lcol is None:
        raise ValueError(f"Text/label columns not found. Columns: {df.columns.tolist()}")

    X = df[[tcol, lcol]].rename(columns={tcol: "text", lcol: "label"}).copy()
    X["text"] = X["text"].astype(str).str.strip()

    # map label to {-1,0,1} if needed
    if X["label"].dtype == "object":
        m = {"positive":1, "pos":1, "1":1, "1.0":1, "true":1,
             "neutral":0, "neu":0, "0":0, "0.0":0,
             "negative":-1, "neg":-1, "-1":-1, "-1.0":-1, "false":-1}
        X["label"] = X["label"].astype(str).str.lower().str.strip().map(m)
    X["label"] = pd.to_numeric(X["label"], errors="coerce")

    # keep valid rows only
    X = X[(X["text"].str.len() > 0) & (X["label"].isin([-1,0,1]))].reset_index(drop=True)
    return X

# If a global X already exists (from earlier cells), reuse it; otherwise load it.
if "X" not in globals():
    X = _load_newdataset_as_X()

# Train/test split
texts = X["text"].astype(str).values
y     = X["label"].astype(int).values
X_train, X_test, y_train, y_test = train_test_split(
    texts, y, test_size=0.30, random_state=42, stratify=y
)

# ---------- 1) Baseline: Count + MultinomialNB ----------
count_vec = CountVectorizer()
Xtr_cnt   = count_vec.fit_transform(X_train)
Xte_cnt   = count_vec.transform(X_test)

nb_cnt   = MultinomialNB()
nb_cnt.fit(Xtr_cnt, y_train)
pred_cnt = nb_cnt.predict(Xte_cnt)

acc_cnt = accuracy_score(y_test, pred_cnt)
print("=== Naive Bayes with Word Frequency (Baseline) ===")
print("Accuracy:", acc_cnt)
print(classification_report(y_test, pred_cnt, digits=4, zero_division=0))

# ---------- 2) Baseline: TF-IDF + MultinomialNB ----------
tfidf_vec = TfidfVectorizer()
Xtr_tfidf = tfidf_vec.fit_transform(X_train)
Xte_tfidf = tfidf_vec.transform(X_test)

nb_tfidf    = MultinomialNB()
nb_tfidf.fit(Xtr_tfidf, y_train)
pred_tfidf  = nb_tfidf.predict(Xte_tfidf)

acc_tfidf = accuracy_score(y_test, pred_tfidf)
print("\n=== Naive Bayes with TF-IDF (Baseline) ===")
print("Accuracy:", acc_tfidf)
print(classification_report(y_test, pred_tfidf, digits=4, zero_division=0))

print("\n=== Quick Baseline Comparison ===")
print(f"TF-IDF accuracy − Count accuracy = {acc_tfidf - acc_cnt:+.4f}")

# ---------- 3) Booster: n-grams + pruning + class weights + alpha search ----------
# Build per-sample weights (inverse of class frequency) to mitigate class imbalance.
unique, counts = np.unique(y_train, return_counts=True)
freq = dict(zip(unique, counts))
sample_weight = np.array([1.0 / freq[c] for c in y_train], dtype=float)

def fit_with_optional_sample_weight(pipe, Xtr, ytr, sw):
    """
    Try to pass sample_weight via pipeline (sklearn 1.2+ supports MultinomialNB(sample_weight)).
    If the environment does not support it, fall back to fitting without weights.
    """
    try:
        pipe.fit(Xtr, ytr, multinomialnb__sample_weight=sw)
    except TypeError:
        pipe.fit(Xtr, ytr)

vectorizers = {
    "COUNT*": CountVectorizer(
        ngram_range=(1, 2),   # unigram + bigram
        min_df=2,             # remove very rare tokens
        max_df=0.90           # remove overly frequent tokens
    ),
    "TFIDF*": TfidfVectorizer(
        ngram_range=(1, 2),
        min_df=2,
        max_df=0.90,
        sublinear_tf=True     # use log(1+tf)
    ),
}
alphas = [1.0, 0.5, 0.3, 0.1]

results = []
best = None

for vname, vect in vectorizers.items():
    for a in alphas:
        pipe = make_pipeline(vect, MultinomialNB(alpha=a))
        fit_with_optional_sample_weight(pipe, X_train, y_train, sample_weight)
        y_pred = pipe.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        results.append({"Vec": vname, "alpha": a, "accuracy": acc, "model": pipe, "y_pred": y_pred})
        if (best is None) or (acc > best["accuracy"]):
            best = {"Vec": vname, "alpha": a, "accuracy": acc, "model": pipe, "y_pred": y_pred}

leaderboard = pd.DataFrame(results).sort_values("accuracy", ascending=False).reset_index(drop=True)
print("\n=== Booster Leaderboard (Top 10) ===")
print(leaderboard.head(10)[["Vec","alpha","accuracy"]])

print("\n=== Best Booster Configuration ===")
print(f"Vectorizer: {best['Vec']} | alpha: {best['alpha']} | accuracy: {best['accuracy']:.4f}")

print("\n=== Classification Report (Best Booster) ===")
print(classification_report(y_test, best["y_pred"], digits=4, zero_division=0))
```

### Phase 2 – Brief Reflection (inefficiencies & improvements)

**Inefficiencies I noticed**
- Multiple full re-vectorizations (Count, TF-IDF) over the whole corpus in separate cells → duplicated work/time.
- Very high-dimensional vocabulary including obvious noise (stock tickers, URLs, stray numbers).
- Sparse pattern-mining I/O (CSV writes/reads) adds overhead and encoding issues on Windows.

**Targeted improvements for this dataset**
- Add a light cleaner: lowercase, remove URLs/mentions, strip repeated punctuation/numbers, keep alphas only.
- Apply a domain stoplist (e.g., “stock”, “market”, “shares”, ticker-like tokens) and min_df ≥ 3 to drop ultra-rare noise.
- Use a single pipeline (`Pipeline([('tfidf', TfidfVectorizer(...)), ('clf', MultinomialNB())])`) to avoid re-fitting twice and keep code tidy.
